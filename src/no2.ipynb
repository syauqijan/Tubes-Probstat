```python
print("Nomor 02 : ")
file = pd.read_csv("anggur.csv")
# ambil kolom numerik
numeric_cols = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides',
                'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']

# plot distribusi untuk setiap kolom numerik
for col in numeric_cols:
    print(f"{col} Penjelasan : ")
    print("----------VISUALISASI DISTRIBUTION PLOT----------")
    # plot histogram
    sns.histplot(file[col], kde=True)
    plt.title(f"Histogram of {col}")
    plt.show()
    
    # plot boxplot
    sns.boxplot(file[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```
