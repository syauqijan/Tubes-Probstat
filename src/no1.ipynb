```python
import pandas as pd
import numpy as np
import scipy.stats as stats

def get_descriptive_statistics(column):
    mean = np.mean(column)
    median = np.median(column)
    mode = stats.mode(column)
    std_dev = np.std(column)
    variance = np.var(column)
    data_range = np.ptp(column)
    minimum = np.min(column)
    maximum = np.max(column)
    quartile1 = np.quantile(column, 0.25)
    quartile2 = np.quantile(column, 0.5)
    quartile3 = np.quantile(column, 0.75)
    iqr = stats.iqr(column)
    skewness = stats.skew(column)
    kurtosis = stats.kurtosis(column)
    
    return pd.DataFrame({
        'mean': mean,
        'median': median,
        'mode': mode.mode,
        'standard deviation': std_dev,
        'variance': variance,
        'range': data_range,
        'minimum': minimum,
        'maximum': maximum,
        'Q1': quartile1,
        'Q2': quartile2,
        'Q3': quartile3,
        'IQR': iqr,
        'skewness': skewness,
        'kurtosis': kurtosis
    })

def describe_column(column_name):
    df = pd.read_csv('../data/anggur.csv')
    column = df[column_name]
    descriptive_statistics = get_descriptive_statistics(column)
    return descriptive_statistics.style.format("{:.2f}")


```
