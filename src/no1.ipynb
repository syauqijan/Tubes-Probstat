```python
import pandas as pd
import scipy.stats as stats

def desc_stat(column_name):
    df = pd.read_csv('../data/anggur.csv')
    
    mean = df[column_name].mean()
    median = df[column_name].median()
    mode = df[column_name].mode()[0]
    std_deviation = df[column_name].std()
    variance = df[column_name].var()
    data_range = df[column_name].max() - df[column_name].min()
    minimum = df[column_name].min()
    maximum = df[column_name].max()
    q1 = df[column_name].quantile(0.25)
    q2 = df[column_name].quantile(0.5)
    q3 = df[column_name].quantile(0.75)
    iqr = q3 - q1
    skewness = df[column_name].skew()
    kurtosis = df[column_name].kurtosis()
    
    desc_stats = {
        'Statistic': ['Mean', 'Median', 'Mode', 'Standard Deviation', 'Variance', 'Range', 'Minimum', 'Maximum', '1st Quartile', '2nd Quartile', '3rd Quartile', 'IQR', 'Skewness', 'Kurtosis'],
        'Value': [mean, median, mode, std_deviation, variance, data_range, minimum, maximum, q1, q2, q3, iqr, skewness, kurtosis]
    }

    df_desc_stats = pd.DataFrame.from_dict(desc_stats)
    df_desc_stats.set_index('Statistic', inplace=True)

    df_desc_stats = df_desc_stats.applymap('{:.3f}'.format)
    
    return df_desc_stats
```

Deskripsi Statistika Kolom Fixed Acidity

```python
desc_stat('fixed acidity')
```
