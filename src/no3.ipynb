```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
import seaborn as sns

# load dataset
print('Nomor 03 : Histogram dan Normality Test')
file = pd.read_csv('../data/anggur.csv')

# extract numeric columns
cols = file.select_dtypes(include=['float64', 'int64']).columns

# loop trough numeric columns
for col in cols:
    #plot histogram
    sns.histplot(file[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.show()
    
    # normality test
    _, pValue = stats.normaltest(file[col])
    alpha = 0.05
    if pValue < alpha:
        print(f'{col} is not normally distributed (p-value = {pValue})')
    else:
        print(f'{col} is normally distributed (p-value = {pValue})')
```
